*Starting simple*

1. Import packages

```python
from sklearn import linear_model
```

2. Create dataset

```python
from sklearn.datasets import make_regression

# Generate a regression dataset with 100 samples and 1 feature
X, y = make_regression(n_samples=200, n_features=1, noise=0.5)

# Print the first 5 samples
print(X[:5])
print(y[:5])

```

```python
# Create a linear regression model
model = linear_model.LinearRegression()

# Train the model
model.fit(X, y)

# Make a prediction
y_pred = model.predict(X)

# Print the first 5 predictions
print(y_pred[:5])

# Print the slope and the intercept and R^2
print("The slope of the linear regression model is:", model.coef_)
print("The intercept of the linear regression model is:", model.intercept_)
print("The R^2 score of the linear regression model is:", model.score(X, y))
```

```python
import matplotlib.pyplot as plt

# plot the regression line
plt.scatter(X, y, color='black')
plt.plot(X, y_pred, color='blue', linewidth=3)
```
